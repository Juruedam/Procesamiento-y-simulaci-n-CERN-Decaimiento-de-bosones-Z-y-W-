```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.optimize import curve_fit
import pandas as pd
from lmfit import Model
```

```python
# Lee el archivo CSV
nombre_archivo_Z11 = 'Zmumu-2011-Open.csv'
df = pd.read_csv(nombre_archivo_Z11, skiprows=0, sep=",", decimal='.')
data_frame_Z11 = df[(df.iloc[:, 2] > 20) & (df.iloc[:, 8] > 20) & (df.iloc[:, 3] > -2.1) & (df.iloc[:, 9] > -2.1)]
array_Z11 = data_frame_Z11.to_numpy()
```

```python
Event_Z11 = array_Z11[:,1]
pt1_Z11 = array_Z11[:,2]
eta1_Z11 = array_Z11[:,3]
phi1_Z11 = array_Z11[:,4]
pt2_Z11 = array_Z11[:,8]
eta2_Z11 = array_Z11[:,9]
phi2_Z11 = array_Z11[:,10]

Inv_mass_Z11 = np.sqrt(2*pt1_Z11*pt2_Z11*(np.cosh(eta1_Z11 - eta2_Z11)-np.cos(phi1_Z11 - phi2_Z11)))
```

```python
bins = np.linspace(np.min(Inv_mass_Z11), np.max(Inv_mass_Z11), 150)  #Personaliza según tus necesidades

def gaussiana(x, amplitude, mean, stddev):
    return amplitude * np.exp(-(x - mean)**2 / (2 * stddev**2))

unique_Z11, counts_Z11 = np.unique(Inv_mass_Z11, return_counts=True)

mean = np.mean(Inv_mass_Z11) + 1.58
std = np.std(Inv_mass_Z11)/3.7

# Crear el histograma
plt.hist(unique_Z11, bins=bins, weights = counts_Z11 , histtype='step',  label= 'Datos Experimentales')

# Gráfica del ajuste
plt.plot(unique_Z11, gaussiana(unique_Z11, 430, mean, std), 'm--', label=f'Ajuste gaussiano\nMean: {mean:.2f}, Std: {std:.2f}')
plt.xlim( 60, 120)
# Configuración de la gráfica
plt.title(r'Masa Invariante mu-mu')
plt.xlabel(r'Masa Invariante [GeV]')
plt.ylabel(r'Eventos')
plt.legend()
plt.savefig('masa_z_exp.png')

# Mostrar el histograma
plt.show()
```

```python
bins = np.linspace(np.min(pt1_Z11), 80, 60)  #Personaliza según tus necesidades

# Crear el histograma
plt.hist(pt1_Z11, bins=bins, histtype='step', label='Datos Experimentales')
plt.xlim(20, 100)

# Configuración de la gráfica
plt.title(r'Momento transversal muon1 (Z)')
plt.xlabel(r'Momento transversal [GeV]')
plt.ylabel(r'Eventos')
plt.legend()
plt.savefig('pt_m1z_exp.png')

# Mostrar la gráfica
plt.show()
```
