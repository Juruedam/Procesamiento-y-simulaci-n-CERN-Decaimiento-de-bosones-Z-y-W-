```python
import numpy as np
import matplotlib.pyplot as plt
import scipy as sp
from scipy.optimize import curve_fit
import pandas as pd
```

```python
# Lee el archivo CSV
nombre_archivo_W11 = 'Wmunu-2011.csv'
data_frame_W11 = pd.read_csv(nombre_archivo_W11, skiprows=0, sep=",", decimal='.')
array_W11 = data_frame_W11.to_numpy() 
```

```python
Event_W11 = array_W11[:,1]
pt_W11 = array_W11[:,2]
eta_W11 = array_W11[:,3]
phi_W11 = array_W11[:,4]
Q_W11 = array_W11[:,5]
chiSq_W11 = array_W11[:,6]
dxy_W11 = array_W11[:,7]
iso_W11 = array_W11[:,8]
MET_W11 = array_W11[:,9]
phiMET_W11 = array_W11[:,10]

print(np.min(pt_W11))

Inv_mass_W11 = np.sqrt(2*pt_W11*MET_W11*(np.cosh(eta_W11)-np.cos(phi_W11-phiMET_W11)))
```

```python
bins = np.linspace(np.min(Inv_mass_W11), 200, 200)  #Personaliza según tus necesidades

# Crear el histograma
plt.hist(Inv_mass_W11, bins=bins, alpha=0.5, label=['Array 1', 'Array 2'], facecolor='red')
# Añadir etiquetas y leyenda
plt.xlabel('Valor')
plt.ylabel('Frecuencia')
plt.legend()

# Mostrar el histograma
plt.show()
```

```python
# Lee el archivo CSV
nombre_archivo_W10 = 'Wmunu-2010.csv'
data_frame_W10 = pd.read_csv(nombre_archivo_W10, skiprows=0, sep=",", decimal='.')
array_W10 = data_frame_W10.to_numpy() 
```

```python
Event_W10 = array_W10[:,1]
pt_W10 = array_W10[:,6]
eta_W10 = array_W10[:,7]
phi_W10 = array_W10[:,8]
Q_W10 = array_W10[:,9]
MET_W10 = array_W10[:,10]
phiMET_W10 = array_W10[:,11]

print(np.max(eta_W10))
Inv_mass_W10 = np.sqrt(2*pt_W10*MET_W10*(np.cosh(eta_W10)-np.cos(phi_W10-phiMET_W10)))
```

```python
bins = np.linspace(np.min(Inv_mass_W10), np.max(Inv_mass_W10), 81)  #Personaliza según tus necesidades

def gaussiana(x, amplitude, mean, stddev):
    return amplitude * np.exp(-(x - mean)**2 / (2 * stddev**2))

unique_W10, counts_W10 = np.unique(Inv_mass_W10, return_counts=True)

mean = np.mean(Inv_mass_W10)-3.7
std = np.std(Inv_mass_W10)/1.18

# Crear el histograma
#plt.hist(Inv_mass_W10, bins=bins, alpha=0.5, label=['Array 1', 'Array 2'], facecolor='green')
plt.hist(unique_W10, bins=bins, weights = counts_W10 , histtype='step',  label= 'Datos Experimentales')

plt.plot(unique_W10, gaussiana(unique_W10, 48, mean, std), 'm--', label=f'Ajuste gaussiano\nMean: {mean:.2f}, Std: {std:.2f}')

# Configuración de la gráfica
plt.title(r'Masa Invariante mu-ne')
plt.xlabel(r'Masa Invariante [GeV]')
plt.ylabel(r'Eventos')
plt.legend()
plt.savefig('masa_w_exp.png')

# Mostrar la gráfica
plt.show()
```
